```
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# load dataset
autolib_df = pd.read_csv('http://bit.ly/autolib_dataset')
autolib_df
```

```
# checking for missing values
autolib_df.isnull().sum().sort_values(ascending=False)
```

```
# check for duplicates in our dataset 
autolib_df.duplicated().value_counts()
```

```
# drop irrelevant columns
autolib_df.drop(["Address", "Geo point", "ID", "Scheduled at", "Displayed comment", "Subscription status"], axis=1, inplace=True)
autolib_df.head() 
```

```
# check for null values
autolib_df.isnull().sum()
```

```
# removing outliers using IQR scores
Q1 = autolib_df.quantile(0.25)
Q3 = autolib_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
autolib_df_clean = autolib_df[~((autolib_df < (Q1 - 1.5 * IQR)) | (autolib_df > (Q3 + 1.5 * IQR)))]
autolib_df_clean.shape

```

```
# sorting the data
autolibdf_sort = autolib_df_clean.sort_values(by=['Public name', 'day', 'hour', 'minute'], ascending=[True, True, True, True])
autolibdf_sort.head(10)

```

```
# find cumulative difference for bluecar counter
autolibdf_sort['Bluecar counter_cummdiff'] = autolibdf_sort['Bluecar counter'].diff(1)
autolibdf_sort['Utilib counter_cummdiff'] = autolibdf_sort['Utilib counter'].diff(1)
autolibdf_sort['Utilib 1.4counter_cummdiff'] = autolibdf_sort['Utilib 1.4 counter'].diff(1)
autolibdf_sort.head()
```

```
# data exploration
# Identifying the highest no. of bluecars that are normally rent out
autolibdf_sort['Bluecar counter'].value_counts().plot(kind='barh', figsize=(5,5))
```

```
# grouping the cummdiff in terms of public name / name of the station
autolibdf_sort['Bluecar counter_cummdiff'] = autolibdf_sort['Bluecar counter'].groupby(autolibdf_sort['Public name']).diff(1)
autolibdf_sort['Utilib counter_cummdiff'] = autolibdf_sort['Utilib counter'].groupby(autolibdf_sort['Public name']).diff(1)
autolibdf_sort['Utilib 1.4counter_cummdiff'] = autolibdf_sort['Utilib counter'].groupby(autolibdf_sort['Public name']).diff(1)
autolibdf_sort
```

```
# dealing with the NaN values
autolibdf_sort['Bluecar counter_cummdiff'] = autolibdf_sort['Bluecar counter_cummdiff'].fillna(0)
autolibdf_sort['Utilib counter_cummdiff'] = autolibdf_sort['Utilib counter_cummdiff'].fillna(0)
autolibdf_sort['Utilib 1.4counter_cummdiff'] = autolibdf_sort['Utilib 1.4counter_cummdiff'].fillna(0)
autolibdf_sort.tail()
```

```
# finding the most popular station overall
most_popularstation = autolibdf_sort['Public name'].value_counts()
most_popularstation
```

```
# the most popular station at the most popular picking hour
mostpop_stat_hour = autolibdf_sort[(autolibdf_sort['hour'] == 21)]['Public name'].value_counts().sort_values(ascending=False)
mostpop_stat_hour
```

The most popular station at the most popular picking hour is Paris/Tronchet/19

```
 # most popular hour of the day for picking up a shared Bluecar in the city of Paris


autolibdf_sort[(autolibdf_sort['City'] == 'Paris') & (autolibdf_sort['Bluecar counter_cummdiff']<0)]['hour'].value_counts().idxmax()
```

The most popular hour for picking up a shared blue car in the city of Paris is 21
